```python
tags = ['P','PC','PCS','PCSB','PCSBI','PCSBIG']

models = ['RF','KNN','LSTM','Ridge']

def load_model_scores(model_name):
    csv_path = RESULTS_PATH / model_name / "scores_tag_all_cv.csv"
    if not csv_path.exists():
        print(f"[Skip] {model_name} scores not found at {csv_path}")
        return pd.DataFrame(columns=["tag","region","model","horizon","mae_mean","mae_std"])
    df = pd.read_csv(csv_path)
    return df

dfs = []
for m in models + ['Naive-12Mean']:
    dfs.append(load_model_scores(m))
df_all = pd.concat(dfs, ignore_index=True) if len(dfs) else pd.DataFrame(columns=["tag","region","model","horizon","mae_mean","mae_std"])

df_all = df_all[df_all["tag"].isin(tags)].copy()

regions = sorted(df_all["region"].dropna().unique().tolist())
horizons_present = sorted(df_all["horizon"].dropna().unique().tolist())

long_horizons = [h for h in horizons_present if h >= 24]
h2i = {h:i for i,h in enumerate(sorted(horizons_present))}

print("Regions:", len(regions), regions[:5], "...")
print("Horizon set:", horizons_present)
print("Long horizons:", long_horizons)

def mean_long_mae(sub):
    if sub.empty:
        return np.nan
    vals = sub.loc[sub["horizon"].isin(long_horizons), "mae_mean"].to_numpy()
    if vals.size == 0:
        return np.nan
    return float(np.mean(vals))

records = []
for mdl in models:
    df_m = df_all[df_all["model"] == mdl]
    for tag in tags:
        df_mt = df_m[df_m["tag"] == tag]
        for region in regions:
            sub = df_mt[df_mt["region"] == region]
            mval = mean_long_mae(sub)
            if not np.isnan(mval):
                records.append({"tag": tag, "model": mdl, "region": region, "mae": mval})

df_long = pd.DataFrame(records)

df_naive = df_all[df_all["model"] == "Naive-12Mean"]
naive_vals = []
for region in regions:
    sub_r = df_naive[(df_naive["region"] == region) & (df_naive["horizon"].isin(long_horizons)) & (df_naive["tag"].isin(tags))]
    if len(sub_r):
        naive_vals.append(float(sub_r["mae_mean"].mean()))
naive_long_baseline = float(np.mean(naive_vals)) if len(naive_vals) else np.nan

for mdl in models:
    print(f"\n=== {mdl} ===")
    dfm = df_long[df_long["model"] == mdl].copy()
    pivot = dfm.pivot(index="region", columns="tag", values="mae").reindex(index=regions, columns=tags)
    summary = dfm.groupby("tag", as_index=True)["mae"].agg(["mean","median"]).reindex(tags)
    summary.columns = ["Mean MAE","Median MAE"]

    display(pivot)
    display(summary)

    (EXPORT_TABLES / f"main_long_{mdl}_region_by_tag.csv").write_text(pivot.to_csv())
    (EXPORT_TABLES / f"main_long_{mdl}_summary.csv").write_text(summary.to_csv())

colors = plt.colormaps["tab10"].colors[:len(tags)]
nrows = 1
ncols = 4
fig, axes = plt.subplots(nrows, ncols, figsize=(18, 4), sharey=True)
axes = axes.flatten()

for ax, mdl in zip(axes, models):
    dfm = df_long[df_long["model"] == mdl].copy()
    mae_lists = [dfm.loc[dfm["tag"] == t, "mae"].values for t in tags]
    bp = ax.boxplot(
        mae_lists,
        labels=tags,
        patch_artist=True,
        showmeans=True,
        meanline=True,
        medianprops={"color":"black","linewidth":1.2},
        meanprops={"color":"#333333","linewidth":1.2}
    )
    for patch, c in zip(bp["boxes"], colors):
        patch.set_facecolor(c)
        patch.set_alpha(0.65)
    if not np.isnan(naive_long_baseline):
        ax.axhline(naive_long_baseline, color="gray", linestyle="-", linewidth=1.3)
    ax.set_title(f"{mdl}", fontsize=11)
    ax.set_xlabel("Tag")
    ax.set_ylabel("MAE")
    ax.tick_params(axis="x", rotation=0)
    plt.ylim(top=12.5)
    plt.tight_layout()

for ax in axes[len(models):]:
    ax.axis("off")

#fig.suptitle("Main Experiment - Long-horizon MAE Distributions by Model and Tag", fontsize=13, y=0.97)
#plt.subplots_adjust(top=0.90, bottom=0.08, left=0.06, right=0.98, hspace=0.45, wspace=0.25)

outfile = EXPORT_FIGS / "main_long_all_models_boxplots.png"
fig.savefig(outfile, bbox_inches="tight")
plt.show()

fig.savefig(outfile, dpi=900, format="png", bbox_inches="tight", pad_inches=0.05)
plt.close(fig)

print("Saved long-horizon pivots, summaries, and consolidated multi-panel figure.")
```

## all models, tags group df

```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

EXPORT_ROOT = Path("paper_exports")
EXPORT_TABLES = EXPORT_ROOT / "tables"
EXPORT_FIGS = EXPORT_ROOT / "figures"
EXPORT_LOGS = EXPORT_ROOT / "logs"

cr = Path("current_run.txt")
RUN_PATH = Path("experiments") / Path(cr.read_text().strip())
CONFIG = json.load(open(RUN_PATH / "config.json"))
RESULTS_PATH = RUN_PATH / "results"

tags = ["P", "PC", "PCS", "PCSB", "PCSBI", "PCSBIG"]

horizon_groups = {
    "Short": [2, 6, 10],
    "Medium": [14, 18, 22],
    "Long": [26, 30, 34],
}

def load_scores(model_name):
    p = RESULTS_PATH / model_name / "scores_tag_all_cv.csv"
    if not p.exists():
        print(f"[Skip] {model_name} scores not found at {p}")
        return pd.DataFrame(columns=["tag", "region", "model", "horizon", "mae_mean", "mae_std"])
    df = pd.read_csv(p)
    return df

df_knn_all = load_scores("KNN")
df_naive_all = load_scores("Naive-12Mean")
df_arima_all = load_scores("ARIMA")

df_knn_all = df_knn_all[df_knn_all["tag"].isin(tags)].copy()
df_naive_all = df_naive_all[df_naive_all["tag"].isin(tags)].copy()
df_arima_all = df_arima_all[df_arima_all["tag"].isin(tags)].copy()

horizons = sorted(df_knn_all["horizon"].dropna().unique().tolist())
regions = sorted(df_knn_all["region"].dropna().unique().tolist())


model_names = ['Ridge','RF','KNN','LSTM','XGB','ARIMA','Naive-12Mean']

records_table = []

for model_name in model_names:
    df_model = load_scores(model_name)
    df_model = df_model[df_model["tag"].isin(tags)].copy()

    available_horizons = sorted(df_model["horizon"].dropna().unique().tolist())

    for group_name, group_horizons in horizon_groups.items():
        group_h = [h for h in group_horizons if h in available_horizons]

        df_group = df_model[df_model["horizon"].isin(group_h)]

        for tag in tags:
            df_tag = df_group[df_group["tag"] == tag]

            vals = df_tag["mae_mean"].values
            mean_val = float(vals.mean())
            std_val = float(vals.std(ddof=0))

            records_table.append(
                {
                    "model": model_name,
                    "group": group_name,
                    "tag": tag,
                    "mae_mean": mean_val,
                    "mae_std": std_val,
                }
            )

df_table_long = pd.DataFrame(records_table)

def format_cell(row):
    return f"{row['mae_mean']:.3f} ± {row['mae_std']:.3f}"

df_table_long["cell"] = df_table_long.apply(format_cell, axis=1)

model_order = ['Ridge','RF','KNN','LSTM','XGB','ARIMA','Naive-12Mean']
group_order = ["Short", "Medium", "Long"]

multi_index = pd.MultiIndex.from_product([model_order, group_order], names=["model", "group"])

df_table_wide = (
    df_table_long.pivot(index=["model", "group"], columns="tag", values="cell")
    .reindex(index=multi_index)
    .reindex(columns=["P", "PC", "PCS", "PCSB", "PCSBI", "PCSBIG"])
)

print("\n=== Table 3 style DataFrame (formatted mean ± std) ===")
display(df_table_wide)

df_table_wide.to_csv(EXPORT_TABLES / "table3_model_group_tag_mae.csv")

df_table_numeric = (
    df_table_long.pivot_table(
        index=["model", "group"],
        columns="tag",
        values=["mae_mean", "mae_std"],
    )
)
df_table_numeric.to_csv(EXPORT_TABLES / "table3_model_group_tag_mae_numeric.csv")

records = []
for tag in tags:
    for h in horizons:
        sub = df_knn_all[(df_knn_all["tag"] == tag) & (df_knn_all["horizon"] == h)]
        records.append(
            {
                "tag": tag,
                "model": "KNN",
                "horizon": h,
                "mae": float(sub["mae_mean"].mean()),
            }
        )
df_h = pd.DataFrame(records)

naive_vals = []
for h in horizons:
    sub = df_naive_all[df_naive_all["horizon"] == h]
    naive_vals.append(float(sub["mae_mean"].mean()) if len(sub) else np.nan)

arima_vals = []
for h in horizons:
    sub = df_arima_all[(df_arima_all["horizon"] == h) & (df_arima_all["tag"] == "P")]
    arima_vals.append(float(sub["mae_mean"].mean()) if len(sub) else np.nan)

print("\n=== KNN per-horizon MAE by tag ===")
df_knn = df_h[df_h["model"] == "KNN"].copy()
wide = df_knn.pivot(index="horizon", columns="tag", values="mae").reindex(index=horizons, columns=tags)
display(wide)
wide.to_csv(EXPORT_TABLES / "main_per_horizon_KNN_by_tag.csv", index=True)

cmap = mpl.colormaps.get_cmap("tab10").resampled(len(tags))
tag_palette = dict(zip(tags, cmap(range(len(tags)))))

fig, ax = plt.subplots(figsize=(8, 5))
final_label_positions = []

for tag in tags:
    sub = df_knn[df_knn["tag"] == tag].sort_values("horizon")
    x = sub["horizon"].values
    y = sub["mae"].values
    ax.plot(x, y, marker="o", markersize=4, linewidth=1.8, color=tag_palette[tag])
    final_label_positions.append((y[-1], tag, tag_palette[tag]))

ax.plot(horizons, naive_vals, linewidth=1.8, color="black", linestyle="--")
if len(naive_vals):
    final_label_positions.append((naive_vals[-1], "Naive-12Mean", "black"))

ax.plot(horizons, arima_vals, linewidth=1.8, color="gray", linestyle="--")
if len(arima_vals):
    final_label_positions.append((arima_vals[-1], "ARIMA (P)", "gray"))

final_label_positions.sort()
y_min, y_max = ax.get_ylim()
gap = (y_max - y_min) * 0.04
adjusted_y = []
for y_val, _, _ in final_label_positions:
    if adjusted_y and y_val - adjusted_y[-1] < gap:
        y_val = adjusted_y[-1] + gap
    adjusted_y.append(y_val)
for i in range(len(adjusted_y) - 2, -1, -1):
    if adjusted_y[i + 1] - adjusted_y[i] < gap:
        adjusted_y[i] = adjusted_y[i + 1] - gap

for (orig_y, label_text, color), adj_y in zip(final_label_positions, adjusted_y):
    ax.text(
        horizons[-1] + 0.6,
        adj_y,
        label_text,
        color=color,
        va="center",
        fontsize=10,
        bbox=dict(boxstyle="round,pad=0.15", facecolor="white", alpha=0.7, edgecolor="none"),
    )

ax.set_xlim(horizons[0], horizons[-1] + 3)
ax.set_xticks(horizons)
ax.set_xticklabels(horizons)
ax.set_title("KNN - Per-horizon MAE Across Tags (with Naive and ARIMA(P) baselines)", fontsize=12)
ax.set_xlabel("Forecast Horizon")
ax.set_ylabel("Mean MAE Across Regions")
ax.grid(alpha=0.4, linestyle="--")
fig.tight_layout()

outfile = EXPORT_FIGS / "main_KNN_per_horizon_lines_labeled.png"
fig.savefig(outfile, bbox_inches="tight")
plt.show()

print("Saved KNN-only per-horizon table, Table 3 style table, and labeled figure with Naive and ARIMA(P) baselines.")
```

## statistical test

```python
from pathlib import Path
import json
import numpy as np
import pandas as pd
from itertools import combinations
from scipy.stats import wilcoxon, friedmanchisquare

EXPORT_ROOT = Path("paper_exports")
EXPORT_TABLES = EXPORT_ROOT / "tables"
EXPORT_FIGS = EXPORT_ROOT / "figures"
EXPORT_LOGS = EXPORT_ROOT / "logs"

EXP_NAME = "additive_tags_main_global_aug"
cr = Path("current_run.txt")
run_rel = cr.read_text().strip()
RUN_PATH = Path("experiments") / Path(run_rel)


CONFIG = json.load(open(RUN_PATH / "config.json"))
RESULTS_PATH = RUN_PATH / "results"

tags = ['P','PC','PCS','PCSB','PCSBI','PCSBIG']
models = ['Ridge','RF','KNN','LSTM','XGB','ARIMA']

def load_scores(model_name: str):
    p = RESULTS_PATH / model_name / "scores_tag_all_cv.csv"
    df = pd.read_csv(p)
    return df

dfs = [load_scores(m) for m in models + ['Naive-12Mean']]
df_all = pd.concat(dfs, ignore_index=True) if len(dfs) else pd.DataFrame(columns=["tag","region","model","horizon","mae_mean","mae_std"])
df_all = df_all[df_all["tag"].isin(tags)].copy()

regions = sorted(df_all["region"].dropna().unique().tolist())
horizons_present = sorted(df_all["horizon"].dropna().unique().tolist())

long_horizons = [h for h in horizons_present if h >= 24]

records = []
for mdl in models:
    df_m = df_all[(df_all["model"] == mdl) & (df_all["horizon"].isin(long_horizons))]
    for tag in tags:
        df_mt = df_m[df_m["tag"] == tag]
        for region in regions:
            sub = df_mt[df_mt["region"] == region]
            if len(sub):
                records.append({"tag": tag, "model": mdl, "region": region, "mae": float(sub["mae_mean"].mean())})
df_long = pd.DataFrame(records)

log_lines = []
summary_stats_all = []

for mdl in models:
    dfm = df_long[df_long["model"] == mdl].pivot(index="region", columns="tag", values="mae")
    dfm = dfm.reindex(index=regions, columns=tags).dropna()
    if dfm.empty or dfm.shape[1] < 2:
        continue
    desc = dfm.describe().loc[["mean","std"]].T.round(4)
    desc.to_csv(EXPORT_TABLES / f"stats_long_{mdl}_mean_std.csv")
    summary_stats_all.append(desc.assign(model=mdl).reset_index().rename(columns={"index":"tag"}))
    stat, p = friedmanchisquare(*[dfm[t].values for t in dfm.columns])
    log_lines.append(f"{mdl} | Friedman p={p:.6f}")
    if p < 0.05:
        pairs = list(zip(tags[:-1], tags[1:]))
        for a, b in pairs:
            if a in dfm.columns and b in dfm.columns:
                w, p2 = wilcoxon(dfm[a], dfm[b])
                p_adj = min(p2 * len(pairs), 1.0)
                sig = "*" if p_adj < 0.05 else ""
                log_lines.append(f"{mdl} | {a} -> {b} | p={p2:.6f} | corrected={p_adj:.6f} {sig}")

best_tags = {}
for mdl in models:
    dfm = df_long[df_long["model"] == mdl]
    best_tags[mdl] = dfm.groupby("tag", as_index=False)["mae"].mean().sort_values("mae").iloc[0]["tag"]

df_best = pd.DataFrame()
for mdl, tag in best_tags.items():
    df_sub = df_long[(df_long["model"] == mdl) & (df_long["tag"] == tag)].copy()
    ser = df_sub.set_index("region")["mae"].reindex(regions)
    df_best = pd.concat([df_best, ser.rename(mdl)], axis=1)

for (m1, m2) in combinations(models, 2):
    if m1 in df_best.columns and m2 in df_best.columns:
        s1 = df_best[m1].dropna()
        s2 = df_best[m2].dropna()
        idx = s1.index.intersection(s2.index)
        w, p2 = wilcoxon(s1.loc[idx], s2.loc[idx])
        log_lines.append(f"Best-tags | {m1} vs {m2} | p={p2:.6f}")

pd.concat(summary_stats_all, axis=0).to_csv(EXPORT_TABLES / "stats_long_all_models_mean_std.csv", index=False)
Path(EXPORT_LOGS / "significance_tests.txt").write_text("\n".join(log_lines), encoding="utf-8")

print("\n".join(log_lines))
print("Saved significance logs and summary tables.")
```
